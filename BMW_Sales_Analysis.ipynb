```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/Data Analisis dan Visualisasi Data Python/BMW sales data (2010-2024) (1).csv'
df = pd.read_csv(path)
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data = df, x='Model', y='Sales_Volume', color = 'Green', estimator = 'sum')
plt.title("BMW Models by Total Sales (2010â€“2024)")
plt.xlabel("Model")
plt.ylabel("Unit Sold")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(data = df, x = 'Year', y = 'Sales_Volume',marker="o")
plt.title("BMW Worldwide Sales (2010â€“2024)")
plt.ylabel("Unit Sold")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=df, x="Year", y="Sales_Volume", hue="Fuel_Type", marker="o")
plt.title("BMW Sales by Fuel Type (2010â€“2024)")
plt.ylabel("Unit Sold")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data = df, x='Sales_Volume', y='Region', color = 'skyblue', estimator = 'sum', orient='h')
plt.title("BMW Sales by Region (2010â€“2024)")
plt.xlabel("")
plt.ylabel("Unit Sold")
plt.tight_layout()
plt.show()
```

```python
df_colorsales = df.groupby("Color", as_index=False)["Sales_Volume"].sum()

plt.figure(figsize=(9,6))
plt.pie(df_colorsales['Sales_Volume'], labels=df_colorsales['Color'], autopct='%1.1f%%')
plt.title('Customer Preference & Market Demand by Car Color (2010â€“2024)')
plt.show()
```
